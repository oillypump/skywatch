```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("ReadParquetDirect")
    .master("spark://spark-master:7077")
    .config(
        "spark.jars",
        "/opt/spark/jars/hadoop-aws-3.3.6.jar,"
        "/opt/spark/jars/aws-java-sdk-bundle-1.12.606.jar,"
        "/opt/spark/jars/hadoop-common-3.3.6.jar,"
        "/opt/spark/jars/iceberg-spark-runtime-3.5_2.12-1.5.0.jar" # Gunakan versi 1.5.0 agar stabil
    )
    .config("spark.hadoop.fs.s3a.endpoint", "http://minio:9000")
    .config("spark.hadoop.fs.s3a.access.key", "minio")
    .config("spark.hadoop.fs.s3a.secret.key", "minio123")
    .config("spark.hadoop.fs.s3a.path.style.access", "true")
    .config("spark.hadoop.fs.s3a.impl", "org.apache.hadoop.fs.s3a.S3AFileSystem")
    # Tambahan penting untuk stabilitas
    .config("spark.hadoop.fs.s3a.aws.credentials.provider", "org.apache.hadoop.fs.s3a.SimpleAWSCredentialsProvider")
    .config("spark.hadoop.fs.s3a.connection.ssl.enabled", "false")
    .getOrCreate()
)
```

```python
from pyspark.sql import SparkSession

# Gabungkan semua JAR yang ada di folder jars
# Ini memastikan Driver & Executor menggunakan file yang sama
spark = (
    SparkSession.builder
    .appName("ReadParquetVerified")
    .master("spark://spark-master:7077")
    # PENTING: Beritahu Executor (Worker) untuk memuat JAR di folder tersebut
    .config("spark.executor.extraClassPath", "/opt/spark/jars/*")
    .config("spark.driver.extraClassPath", "/opt/spark/jars/*")
    # Daftarkan JAR secara eksplisit untuk dikirim lewat network jika perlu
    .config("spark.jars", "/opt/spark/jars/hadoop-common-3.3.6.jar,"
                          "/opt/spark/jars/hadoop-aws-3.3.6.jar,"
                          "/opt/spark/jars/aws-java-sdk-bundle-1.12.606.jar,"
                          "/opt/spark/jars/iceberg-spark-runtime-3.5_2.12-1.5.0.jar")
    # Konfigurasi S3A MinIO
    .config("spark.hadoop.fs.s3a.endpoint", "http://minio:9000")
    .config("spark.hadoop.fs.s3a.access.key", "minio")
    .config("spark.hadoop.fs.s3a.secret.key", "minio123")
    .config("spark.hadoop.fs.s3a.path.style.access", "true")
    .config("spark.hadoop.fs.s3a.impl", "org.apache.hadoop.fs.s3a.S3AFileSystem")
    .config("spark.hadoop.fs.s3a.aws.credentials.provider", "org.apache.hadoop.fs.s3a.SimpleAWSCredentialsProvider")
    # Matikan prefetch untuk menghindari bug statistik yang tadi
    .config("spark.hadoop.fs.s3a.prefetch.enabled", "false")
    .getOrCreate()
)
```

```python
# Coba baca lagi
df = spark.read.parquet("s3a://raw/yellow_tripdata_2023-01.parquet")
df.show(5)
```

```python
df = spark.read.parquet(
    "s3a://raw/yellow_tripdata_2023-01.parquet"
)

df.printSchema()
df.show(5, truncate=False)
```
